### 1.1.2 Agrupación de datos:

**TABLA TOTAL:**

```python
df_total = pd.concat([df1, df2, df3, df4], ignore_index = True)
```

## 1.2 Agrupación de datos:

Antes de subir las tablas a la **Base de Datos**, vamos a generar los DataFrames en los que agrupar los datos.

### 1.2.1 Tablas sin Foreign Keys:

#### 1.2.1.1 Vertical:

**Tabla X**: Vertical

Distintos Bootcamps que ofrece The Bridge.

```python
df_vertical = pd.DataFrame({
    "bootcamp": df_claustro["Vertical"].unique()
})
df_vertical.reset_index(names = "id_vertical", inplace = True)
df_vertical
```

#### 1.2.1.2 Promoción:

```python
df_promo = df_total[["Promoción", "Fecha_comienzo"]].drop_duplicates().reset_index(drop = True)
df_promo.reset_index(names = "id_promocion", inplace = True)
df_promo
```

#### 1.2.1.3 Campus:

```python
 df_campus = pd.DataFrame({
    "ubicacion": df_claustro["Campus"].unique(),
})
df_campus.reset_index(names = "id_campus", inplace = True)
df_campus
```

#### 1.2.1.4 Proyecto:

```python
df_proyecto = pd.DataFrame({
    "nombre_proyecto": [elem for elem in df_total.columns if "Proyecto" in elem],
})
df_proyecto.reset_index(names = "id_proyecto", inplace = True)
df_proyecto
```

#### 1.2.1.5 Profesorado:

```python
df_profesorado = df_claustro[["Nombre", "Rol"]]
df_profesorado.reset_index(names = "id_profesor", inplace = True)
df_profesorado
```

### 1.2.2 Tablas con Foreign Keys:

#### 1.2.2.1 Clase:

Partiremos del DataFrame df_claustro (creado a partir del fichero *claustro.csv*). Primero, haremos merge con el DataFrame del vertical:

```python
df_claustro
```

```python
df_vertical
```

```python
a = pd.merge(df_claustro, df_vertical, left_on = "Vertical", right_on = "bootcamp")
#a = a.drop("Vertical", axis = 1)
a = a.drop(["Vertical", "bootcamp"], axis = 1)
a
```

Después, haremos merge con el DataFrame del profesorado:

```python
b = pd.merge(a, df_profesorado, on = "Nombre")
b.drop(["Nombre", "Rol_x", "Rol_y"], axis = 1, inplace = True)
b
```

A continuación, haremos merge con el DataFrame del campus:

```python
df_campus
```

```python
c = pd.merge(b, df_campus, left_on = "Campus", right_on = "ubicacion")
c.drop(["Campus", "ubicacion"], axis = 1, inplace = True)
c
```

**Finalmente**, haremos merge con el DataFrame de la promoción:

```python
df_promo
```

```python
d = pd.merge(c, df_promo, left_on = "Promocion", right_on = "Promoción")
d.drop(["Promocion", "Promoción", "Fecha_comienzo"], axis = 1, inplace = True)
d
```

```python
d
```

```python
d = d[["id_campus", "id_promocion", "id_vertical", "Modalidad"]] # Reordenamos las columnas

# Creamos un diccionario para cambiar el nombre de las columnas
rename = {
    "id_campus": "campus_id",
    #"id_profesor": "profesor_id",
    "id_promocion": "promocion_id",
    "id_vertical": "vertical_id"
}

df_clase = d.rename(mapper = rename, axis = 1)


df_clase = df_clase.reset_index(names = "id_clase")
df_clase
```

```python
df_clase = df_clase.drop_duplicates(["campus_id", "promocion_id", "vertical_id"])
df_clase
```

#### 1.2.2.2 Cuadrante:

Hay que usar merge con las tablas **Profesorado** y **Clase**.

```python
df_claustro
```

```python
a = pd.merge(df_claustro, df_vertical, left_on = "Vertical", right_on = "bootcamp").drop(["bootcamp", "Vertical"], axis = 1)
a
```

```python
b = pd.merge(a, df_promo, left_on = "Promocion", right_on = "Promoción").drop(["Promocion", "Promoción"], axis = 1)
b
```

```python
c = pd.merge(b, df_campus, left_on = "Campus", right_on = "ubicacion").drop(["Campus", "ubicacion", "Rol", "Fecha_comienzo"], axis = 1)
c
```

```python
d = pd.merge(c, df_profesorado, on = "Nombre").drop(["Rol", "Nombre", "Modalidad"], axis = 1)
d
```

```python
e = pd.merge(d, df_clase, left_on = ["id_vertical", "id_promocion", "id_campus"], right_on = ["vertical_id", "promocion_id","campus_id"])
e
```

```python
drops = [elem for elem in e.columns if elem not in {"id_profesor", "id_clase"}]
e = e.drop(drops, axis = 1)
e
```

```python
df_cuadrante = e.reset_index(names = "id_cuadrante")
df_cuadrante
```
