# Tratamiento de datos

## Carga de datos:

Disponemos de 4 ficheros csv que contienen los datos con los que construir la base de datos:
- `clase_1.csv` y `clase_2.csv` contienen los datos de los alumnos pertenecientes al bootcamp de *Data Science*.
- `clase_3.csv` y `clase_3.csv` contienen los datos de los alumnos pertenecientes al bootcamp de *Full Stack*.
- `claustro.csv` contiene los datos del claustro de profesores.

Echemos un vistazo al contenido de los csv:

```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/clase_1.csv', sep = ';').assign(clase_id=1, vertical='DS', vertical_id=0)
df2 = pd.read_csv('data/clase_2.csv', sep = ';').assign(clase_id=2, vertical='DS', vertical_id=0)
df3 = pd.read_csv('data/clase_3.csv', sep = ';').assign(clase_id=3, vertical='FS', vertical_id=1)
df4 = pd.read_csv('data/clase_4.csv', sep = ';').assign(clase_id=4, vertical='FS', vertical_id=1)
df_claustro = pd.read_csv("data/claustro.csv", sep = ';')
```

```python
df1.head(3)
```

```python
df_claustro
```

### Tabla 'Maestra' Total

```python
df_total = pd.concat([df1, df2, df3, df4], ignore_index = True)
```

```python
df_total.head(3)
```

```python
df_total['promocion_id'] = df_total['Promoción'].map({'Septiembre': 0, 'Febrero': 1})
df_total['campus_id'] = df_total['Campus'].map({'Madrid': 0, 'Valencia': 1})
```

```python
df_total.head(3)
```

```python
df_total.info()
```

### Tabla Clase

```python
df_clase = pd.DataFrame({
    "id_clase": df_total["clase_id"],
    "campus_id": df_total["campus_id"],
    "promocion_id": df_total["promocion_id"],
    "vertical_id": df_total["vertical_id"]
})
```

```python
df_clase = df_clase.drop_duplicates()
```

```python
df_clase
```

### Tabla Alumno

```python
df_alumno = pd.DataFrame({
    "clase_id": df_total["clase_id"],
    "nombre": df_total["Nombre"],
    "email": df_total["Email"]
})
```

```python
df_alumno["id_alumno"] = range(len(df_alumno))
#df_alumno = df_alumno.set_index("id_alumno")
df_alumno = df_alumno[['id_alumno', 'clase_id', 'nombre', 'email']]
```

```python
df_alumno
```

### Tabla Vertical:

```python
df_total.head(3)
```

```python
df_vertical = pd.DataFrame({
    "bootcamp": df_total["vertical"].unique(),
    "id_vertical": df_total["vertical_id"].unique(),
})
```

```python
df_vertical['id_vertical'] = df_vertical.index
#df_vertical = df_vertical.set_index('id_vertical')
df_vertical = df_vertical[['id_vertical', 'bootcamp']]
df_vertical
```

### Tabla Promoción

```python
df_promocion = pd.DataFrame({
    "id_promocion": df_total["promocion_id"],
    "nombre": df_total["Promoción"]
})
df_promocion = df_promocion.drop_duplicates(subset = "nombre")
#df_promocion = df_promocion.set_index('id_promocion')
df_promocion
```

### Tabla Campus

```python
df_campus = pd.DataFrame({
    "id_campus": df_total["campus_id"],
    "ubicación": df_total["Campus"]
})

df_campus = df_campus.drop_duplicates(subset = "ubicación")
#df_campus = df_campus.set_index('id_campus')
df_campus
```

### Tabla Profesorado:

```python
df_profesorado = pd.DataFrame({
    "nombre": df_claustro["Nombre"].unique(),
    'rol': df_claustro["Rol"]
})
```

```python
df_profesorado['id_profesor'] = df_profesorado.index
#df_profesorado = df_profesorado.set_index('id_profesor')
df_profesorado = df_profesorado[['id_profesor', 'nombre', 'rol']]
df_profesorado
```

### Tabla Cuadrante

```python
df_claustro = df_claustro.reset_index().rename(columns={'index': 'id_profesor'})
df_claustro
```

```python
df_profe_clase = df_claustro.merge(    
    df_total[['clase_id', 'vertical_id', 'vertical', 'promocion_id', 'Promoción', 'campus_id', 'Campus']].drop_duplicates(),
    left_on=['Vertical', 'Promocion', 'Campus'],
    right_on=['vertical', 'Promoción', 'Campus'],
    how='inner')

df_profe_clase
```

```python
df_cuadrante = df_profe_clase[['id_profesor', 'clase_id']].copy()
df_cuadrante['id_cuadrante'] = range(1, len(df_cuadrante) + 1)
df_cuadrante = df_cuadrante[['id_cuadrante', 'id_profesor', 'clase_id']]
df_cuadrante = df_cuadrante.rename(columns={'id_profesor': 'profesor_id'})
df_cuadrante
```

### Tabla Proyecto

```python
df_total.head(1)
```

```python
columnas_proyectos = []

for col in df_total.columns:
    if col.startswith('Proyecto_'):
        col = col.replace('Proyecto_', '')
        columnas_proyectos.append(col)

print(columnas_proyectos)
```

```python
df_proyecto = pd.DataFrame({
    "nombre_proyecto": columnas_proyectos
})

df_proyecto["id_proyecto"] = range(len(df_proyecto))
cambio_orden = ['id_proyecto','nombre_proyecto']
df_proyecto = df_proyecto[cambio_orden]
df_proyecto
```

### Tabla Notas

```python
df_total.head(2)
```

```python
# Lista de columnas con notas (con prefijo)
proyecto_cols = [f'Proyecto_{nombre}' for nombre in df_proyecto['nombre_proyecto']]
proyecto_cols
```

```python
proyecto_cols = [col for col in df_total.columns if col.startswith('Proyecto_')]

notas_list = []

for idx, fila in df_total.iterrows():
    id_alumno = idx
    for proyecto_col in proyecto_cols:
        calificacion = fila[proyecto_col]
        if pd.notna(calificacion):
            nombre_proyecto = proyecto_col.replace('Proyecto_', '')
            id_proyecto = df_proyecto.loc[df_proyecto['nombre_proyecto'] == nombre_proyecto, 'id_proyecto'].values[0]
            
            notas_list.append({
                'alumno_id': id_alumno,
                'proyecto_id': id_proyecto,
                'calificacion': calificacion
            })

df_notas = pd.DataFrame(notas_list)

df_notas['id_nota'] = range(1, len(df_notas) + 1)

df_notas = df_notas[['id_nota', 'alumno_id', 'proyecto_id', 'calificacion']]

df_notas['calificacion'] = df_notas['calificacion'].map({'Apto': True, 'No Apto': False})

df_notas
```

## Guardamos las tablas como CSV

```python
df_clase.to_csv("data/clase.csv", index = False)
df_alumno.to_csv("data/alumno.csv", index = False)
df_vertical.to_csv("data/vertical.csv", index = False)
df_promocion.to_csv("data/promocion.csv", index = False)
df_campus.to_csv("data/campus.csv", index = False)
df_profesorado.to_csv("data/profesorado.csv", index = False)
df_cuadrante.to_csv("data/cuadrante.csv", index = False)
df_proyecto.to_csv("data/proyecto.csv", index = False)
df_notas.to_csv("data/notas.csv", index = False)
```
